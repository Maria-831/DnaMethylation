### What does the distribution of DNA Methylation look like for each chromosome for each group? 

In this section we see what DNA Methylation distribution looks like for each chromosome for each group. First I obtained the average beta value for each patient, from group 1, at the specific chromosome and graph the distribution. Then I did the same for group 2.


```python
def chrdata(a,b,c,d,e,f,g,h,i,j,k):
    
    #Chromosome Analysis Function. 
    #This fuction takes 10 datasets and 1 chromosome of interest.
    
    #This function outputs the mean of the beta value at the chromosome
    #of interest for the whole group.
    
    #So, if given the 10 patient's data from group one, 
    #it outputs the average beta value of 
    #the group at the specified chromosome.
    
    #gp1chr stands for group, patient 1, chromosome. 
    #k is the chromosome of interest.
    
    gp1chr = a.loc[a['Chromosome'] == k] 
    gp2chr = b.loc[b['Chromosome'] == k]
    gp3chr = c.loc[c['Chromosome'] == k]
    gp4chr = d.loc[d['Chromosome'] == k]
    gp5chr = e.loc[e['Chromosome'] == k]
    gp6chr = f.loc[f['Chromosome'] == k]
    gp7chr = g.loc[g['Chromosome'] == k]
    gp8chr = h.loc[h['Chromosome'] == k]
    gp9chr = i.loc[i['Chromosome'] == k]
    gp10chr = j.loc[j['Chromosome'] == k]
    
    #Creates a dataset containing all beta values 
    #and the last column is the mean of each row.
    data = pd.DataFrame()
    
    #Stores the beta value column extracted from the main dataset 
    #that belongs to the specified chromosome. 
    
    data['gp1c_Beta_Value'] = gp1chr['Beta_value'].copy()
    data['gp2c_Beta_Value'] = gp2chr['Beta_value'].copy()
    data['gp3c_Beta_Value'] = gp3chr['Beta_value'].copy()
    data['gp4c_Beta_Value'] = gp4chr['Beta_value'].copy()
    data['gp5c_Beta_Value'] = gp5chr['Beta_value'].copy()
    data['gp6c_Beta_Value'] = gp6chr['Beta_value'].copy()
    data['gp7c_Beta_Value'] = gp7chr['Beta_value'].copy()
    data['gp8c_Beta_Value'] = gp8chr['Beta_value'].copy()
    data['gp9c_Beta_Value'] = gp9chr['Beta_value'].copy()
    data['gpc10_Beta_Value'] = gp10chr['Beta_value'].copy()

    #For each row, the mean is calculated and stored in 
    #a new column named "GMeanC_Beta_Value".
    data['GMeanC_Beta_Value'] =  data.apply(lambda x: x.mean(), axis=1)
    
    #Oputputs only the mean column.
    return data['GMeanC_Beta_Value']
```

```python
#Lets use the function to extract 
#all beta values from each patient in group 1 for Chromosome 1.
group1chr1 = chrdata(g1p1,g1p2,g1p3,g1p4,g1p5,
                     g1p6,g1p7,g1p8,g1p9,g1p10,'chr1')

#Lets use the function to extract 
#all beta values from each patient in group 2 for Chromosome 1.
group2chr1 = chrdata(g2p1,g2p2,g2p3,g2p4,g2p5,
                     g2p6,g2p7,g2p8,g2p9,g2p10,'chr1')
```

### Which group shows a significantly different distribution of DNA Methylation levels at a specific chromosome?

In this section the Student t-test for independent samples is used to calculate the hypotheses. 

**Null Hypothesis:** The distribution of beta values within each chromosome of patients from both group is the **same.**

**Alternative Hypothesis:** The distribution of beta values within each chromosome of patients from both group is **different.**

Below we will see if DNA methylation is present at different levels in chromosomes of each patient of both groups. (e.g. is chromosome 1 significantly hypermethylated (beta value > 0.7) more frequently in smokers than non-smokers?)

The samples are being drawn from a Gaussian distribution and so student t-tests can be performed. 

```python
#Writing out the student t-test

#Necesary packages
from scipy.stats import t
import numpy as np
import math

#Creating a function called chrsig = chromosome significance.

#It takes 2 datasets, a and b, 
#and calculates their statistical significance.

def chrsig(a,b):
    #Lets get the means
    avg1 = a.mean()
    avg2 = b.mean()

    #Standard deviations
    stdv1 = np.std(a, ddof=1)
    stdv2 = np.std(b, ddof=1)

    #Standard errors
    N1 = len(a) 
    N2 = len(b)
    
    se1 = stdv1/math.sqrt(N1) 
    se2 = stdv2/math.sqrt(N2)

    #standard error of differences
    sed = math.sqrt(se1**2.0 + se2**2.0)

    #t-statistic
    t_stats = (avg1 - avg2) / sed

    #degrees of freedom
    dof = N1 + N2 - 2
    
    #Alpha
    a = 0.05

    #p-value
    p = (1 - t.cdf(abs(t_stats), dof)) * 2


    #Result/Output
    if p > a:
        print('Distributions are the same. Accept null hypothesis. The p-value is %.4f' 
              % p)
    else:
        print('Distributions are different. Reject the null hypothesis. The p-value is %.4f' 
              % p)
    
```
